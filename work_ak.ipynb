```python
df_bed_strength = df_hospital_bed[['hospital_id', 'bed_strength']]

print(df_bed_strength.head())

df_private_helper['admission_dt'] = pd.to_datetime(df_private_helper['admission_dt'])

df_private_helper['admission_date'] = df_private_helper['admission_dt'].dt.date


daily_admissions = (
    df_private_helper
    .groupby(['hospital_id'], as_index=False)
    .agg(daily_admissions=('registration_id', 'count'))
)

df_merged = daily_admissions.merge(
    df_hospital_bed[['hospital_id', 'bed_strength']],
    on='hospital_id',
    how='left'
)

df_suspicious = df_merged[
    (df_merged['bed_strength'].notna()) &
    (df_merged['daily_admissions'] > df_merged['bed_strength'])
]

df_suspicious['excess_admissions'] = (
    df_suspicious['daily_admissions'] - df_suspicious['bed_strength']
)

# df_suspicious = df_suspicious.sort_values(
#     ['admission_date', 'hospital_id']
# )

df_suspicious
```

```python
import os
import django

os.environ.setdefault(
    "DJANGO_SETTINGS_MODULE",
    "pmjay_fraud_dashboard.settings"  # change this
)

django.setup()

from datetime import timedelta
from django.db.models import Q, Sum
from pmjay_fraud_dashboard_app.views import parse_date
from pmjay_fraud_dashboard_app.views import high_value_claims_base_query

class MockRequest:
    def __init__(self, params: dict):
        self.GET = params


def get_high_value_claims_notebook(request):
    district_param = request.GET.get('district', '')
    districts = district_param.split(',') if district_param else []

    startDate = request.GET.get('start_date')
    endDate = request.GET.get('end_date')

    # PRESERVED date parsing logic
    start_date, end_date = parse_date(startDate, endDate)

    # Base queryset
    cases = high_value_claims_base_query(start_date, end_date)
    print(f"Total cases in date range: {cases.count()}")

    if districts:
        cases = cases.filter(patient_district_name__in=districts)

    # Time thresholds
    yesterday = end_date - timedelta(days=1)
    thirty_days_ago = end_date - timedelta(days=30)

    # Surgical cases (≥ 1,00,000)
    surgical_cases = cases.filter(
        Q(case_type__iexact='SURGICAL'),
        Q(amount_claim_initiated__gte=100000)
    )

    # Medical cases (≥ 25,000)
    medical_cases = cases.filter(
        Q(case_type__iexact='MEDICAL'),
        Q(amount_claim_initiated__gte=25000)
    )

    surgical_total = surgical_cases.aggregate(
        Sum('amount_claim_initiated')
    )['amount_claim_initiated__sum'] or 0

    medical_total = medical_cases.aggregate(
        Sum('amount_claim_initiated')
    )['amount_claim_initiated__sum'] or 0

    data = {
        'total_count': surgical_cases.count() + medical_cases.count(),

        'yesterday_count': (
            surgical_cases.filter(preauth_init_date__date=yesterday).count() +
            medical_cases.filter(preauth_init_date__date=yesterday).count()
        ),

        'last_30_days_count': (
            surgical_cases.filter(preauth_init_date__date__gte=thirty_days_ago).count() +
            medical_cases.filter(preauth_init_date__date__gte=thirty_days_ago).count()
        ),

        'surgical': {
            'count': surgical_cases.count(),
            'amount': surgical_total,
            'yesterday': surgical_cases.filter(preauth_init_date__date=yesterday).count(),
            'last_30_days': surgical_cases.filter(
                preauth_init_date__date__gte=thirty_days_ago
            ).count()
        },

        'medical': {
            'count': medical_cases.count(),
            'amount': medical_total,
            'yesterday': medical_cases.filter(preauth_init_date__date=yesterday).count(),
            'last_30_days': medical_cases.filter(
                preauth_init_date__date__gte=thirty_days_ago
            ).count()
        }
    }

    return data

mock_request = MockRequest({
    "start_date": "2025-10-01",
    "end_date": "2025-10-31",
    "district": ""
})

result = get_high_value_claims_notebook(mock_request)
result
```

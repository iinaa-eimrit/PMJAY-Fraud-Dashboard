```python
def calculate_ot_violations(df_cases, df_hospitals):
    """
    Replicates OT overload logic exactly as in load_dataframes().

    Parameters
    ----------
    df_cases : DataFrame
        From pmjay_fraud_dashboard_app_last24hour
        Must contain:
          - hospital_id
          - preauth_init_date (datetime)
    
    df_hospitals : DataFrame
        From pmjay_fraud_dashboard_app_hospital_beds
        Must contain:
          - hospital_id
          - number_of_surgeons

    Returns
    -------
    df_cases_with_ot : DataFrame
        Original df_cases with a new boolean column 'ot_violation'
    """

    df = df_cases.copy()

    # -----------------------------------
    # Step 1: Build capacity map
    # capacity = number_of_surgeons * 30
    # -----------------------------------
    capacity_map = (
        df_hospitals
        .dropna(subset=['number_of_surgeons'])
        .assign(capacity=lambda x: x['number_of_surgeons'] * 30)
        .set_index('hospital_id')['capacity']
        .to_dict()
    )

    # -----------------------------------
    # Step 2: Sort cases by hospital + time
    # -----------------------------------
    df['preauth_init_date'] = pd.to_datetime(
        df['preauth_init_date'],
        format='mixed',
        errors='coerce'
    )

    # df = df.sort_values(
    #     by=['hospital_id', 'preauth_init_date'],
    #     ascending=[True, True]
    # )

    df = df.sort_values(by=['preauth_init_date'], ascending=True)
    df_ot_case = df.groupby('hospital_id')
    # -----------------------------------
    # Step 3: Running count per hospital
    # -----------------------------------
    df_ot_case['hospital_running_count'] = (
        df.groupby('hospital_id').cumcount() + 1
    )

    # -----------------------------------
    # Step 4: Flag OT violations
    # -----------------------------------
    def is_ot_violation(row):
        cap = capacity_map.get(row['hospital_id'])
        if cap is None:
            return False
        return row['hospital_running_count'] > cap

    df['ot_violation'] = df.apply(is_ot_violation, axis=1)

    return df
```

```python
# Load hospital beds table
hospital_df = pd.read_sql_query(
    'SELECT * FROM pmjay_fraud_dashboard_app_hospitalbeds',
    conn
)

# Apply OT logic
df_with_ot = calculate_ot_violations(df1, hospital_df)

# Inspect violations
df_with_ot[df_with_ot['ot_violation'] == True]
```
